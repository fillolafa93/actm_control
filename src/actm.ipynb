```python
import numpy as np
import matplotlib.pyplot as plt

# %% Model Parameters
N = 20               # Number of sections
K = 360 * 2          # Time horizon (2 hours) in steps
T = (10 / 3600)      # Sampling time interval (10 seconds in hours)

# On-ramp indicator: Ir=1 if there is an onramp in that section.
Ir = np.zeros(N)
Ir[4] = 1   # MATLAB: Ir(5)=1 (Python indices start at 0)
Ir[9] = 1   # MATLAB: Ir(10)=1
Ir[11] = 1  # MATLAB: Ir(12)=1

# Section lengths (all sections 0.5 corresponding to 500 m)
L = np.full(N, 0.5)

# Maximum densities
rho_max = np.full(N, 300)

# Free-flow speeds and congested wave speeds
vf = np.full(N, 100)
w = np.full(N, 27)

# %% Initial Conditions
# Initialize density and flow in all sections.
rho_iniz = np.full(N, 60)
fi_iniz = rho_iniz * 100

# Boundary conditions for the first section (for all time steps)
rho_sez0 = np.full(K, 60)
fi_sez0 = rho_sez0 * 100

# Boundary conditions for the last section (for all time steps)
rho_sezfin = np.full(K, 80)
fi_sezfin = rho_sezfin * 100

# %% ON-ramp Demand
# Create a demand matrix with dimensions (N, K)
dem = np.zeros((N, K))

# For the first 100 time steps, set high demand on on-ramps in sections 5, 10, 12.
for k in range(100):
    dem[4, k] = 2000   # Section 5 (MATLAB index 5 -> Python index 4)
    dem[9, k] = 2000   # Section 10
    dem[11, k] = 2000  # Section 12

# For time steps 101 to K, assign lower demand.
for k in range(100, K):
    dem[4, k] = 100    # Section 5
    dem[9, k] = 100    # Section 10
    dem[11, k] = 50    # Section 12

# %% No Control Simulation

# Allocate arrays for densities and queue lengths (queue lengths are not used further)
rho = np.zeros((N, K + 1))
ql = np.zeros((N, K + 1))

# Traffic volumes at on-ramps (here simply a copy of the demand)
r = dem.copy()

# Allocate interface flow array
fi = np.zeros((N, K + 1))

# Initial density calculation (at time index 0)
for i in range(N - 1):
    # For sections 1 to N-1 (adjusted for Python 0-indexing)
    rho[i, 0] = rho_iniz[i] + T / L[i] * (fi_iniz[i] - fi_iniz[i + 1])
# For the last section:
rho[N - 1, 0] = rho_iniz[N - 1] + T / L[N - 1] * (fi_iniz[N - 1] - vf[N - 1] * rho_iniz[N - 1])

# Time stepping loop
for k in range(K):
    # Compute interface flow for the first section using the boundary condition.
    fi[0, k] = min(fi_sez0[k], w[0] * (rho_max[0] - rho[0, k]) - r[0, k])
    
    # For sections 2 to N, compute the interface flow.
    for i in range(1, N):
        fi[i, k] = min(vf[i - 1] * rho[i - 1, k], w[i] * (rho_max[i] - rho[i, k]) - r[i, k])
    
    # Update densities for sections 1 to N-1
    for i in range(N - 1):
        rho[i, k + 1] = rho[i, k] + T / L[i] * (fi[i, k] + r[i, k] - fi[i + 1, k])
    
    # For the last section, adjust the boundary flow and update density.
    fi_sezfin[k] = min(vf[N - 1] * rho[N - 1, k], fi_sezfin[k])
    rho[N - 1, k + 1] = rho[N - 1, k] + T / L[N - 1] * (fi[N - 1, k] + r[N - 1, k] - fi_sezfin[k])

# Plot the density as a mesh (using imshow for visualization)
plt.figure()
plt.imshow(rho, aspect='auto', origin='lower')
plt.colorbar(label='Density')
plt.xlabel('Time Step')
plt.ylabel('Section')
plt.title('Traffic Density Over Time and Sections')
plt.show()
```
